```python
def candidate_elimination(examples,target):
    specific_h = examples[0][:-1]
    general_h = [['?' for _ in range(len(specific_h))] for _ in range(len(specific_h))]

    for example in examples:
        if example[-1] == target:
            for i in range(len(specific_h)):
                if example[i] != specific_h[i]:
                    specific_h[i] = '?'
                    general_h[i][i] = '?'
        else:
            for i in range(len(specific_h)):
                if example[i] != specific_h[i]:
                    general_h[i][i] = specific_h[i]
                else:
                    general_h[i][i] = '?'

    return specific_h, general_h
```

```python
# Example dataset
dataset = [
    ['Overcast','Hot','High','False','Yes'],
    ['Rainy','Mild','High','False','Yes'],
    ['Rainy','Cool','Normal','False','Yes'],
    ['Rainy','Cool','Normal','True','No'],
    ['Overcast','Cool','Normal','True','Yes'],
    ['Sunny','Mild','High','False','No'],
    ['Sunny','Cool','Normal','False','Yes'],
    ['Rainy','Mild','Normal','False','Yes'],
    ['Sunny','Mild','Normal','True','Yes'],
    ['Rainy','Mild','High','True','No']
]
```

```python
#Result
countSpecificYes=0
countSpecificNo=0
countGeneralYes=0
countGeneralNo=0

if(dataset[-1]=='Yes'):
  specificY, generalY = candidate_elimination(dataset,'Yes')
  for i in specificY:
    if(i=='?'):
      countSpecificYes=countSpecificYes+1
  for i in generalY:
    if(i=='?'):
      countGeneralYes=countGeneralYes+1
else:
  specificN, generalN = candidate_elimination(dataset,'No')
  for i in specificN:
    if(i=='?'):
      countSpecificNo=countSpecificNo+1
  for i in generalN:
    if(i=='?'):
      countGeneralNo=countGeneralNo+1

if(countSpecificYes>=countSpecificNo and countGeneralYes>=countSpecificNo):
  print("Specific Hypothesis:", specificY)
  print("General Hypothesis:")
  for hypothesis in generalY:
    print(hypothesis)
else:
  print("Specific Hypothesis:", specificN)
  print("General Hypothesis:")
  for hypothesis in generalN:
    print(hypothesis)
```
